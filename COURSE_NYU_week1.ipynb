```
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
#-------------------------------------------------------------------#
SMALL_SIZE = 12
MEDIUM_SIZE = 14
BIGGER_SIZE = 14

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE, titleweight='bold')     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE, labelweight='bold')    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
#-------------------------------------------------------------------#
```

```
# generate some points in 2-D space
import random as r
X=np.random.randn(1000,2)
#X=np.random.rand(1000,2)
color=X[:,0]
```

# Linear transformations: rotation, compression, stretch, reflexion

```
#visualizing linear transformations
fig=plt.figure(figsize=(10,5))
plt.subplot(2,5,1)
plt.scatter(X[:,0],X[:,1],c=color)
plt.title('Original')

for i in range(9):
  W=np.random.randn(2,2) #generate random matrix
  Y=np.dot(X,W) #linear transformation Y=XW
  U, S, V = np.linalg.svd(W) #SVD
  plt.subplot(2,5,i+2)
  plt.scatter(Y[:,0],Y[:,1],c=color)
  plt.title( f"S: {np.round(S[0],1)},{np.round(S[1],1)} ")
  plt.subplots_adjust(wspace=.6, hspace=0.6)
```



*   Large singular values: stretch the points
*   Small singular values: squish them together
*   s~0 kills the associated dimension






Linear transforms can rotate, reflect, stretch and compress, but cannot curve
We need non-linearities for this.

```
#visualizing linear transformations
fig=plt.figure()
plt.subplot(2,3,1)
plt.scatter(X[:,0],X[:,1],c=color)
plt.axis('off')
plt.title('Original')

for i in range(5):
  model=Sequential()
  model.add(Dense(2,input_dim=X.shape[1],activation='linear'))
  model.add(Dense(2,activation='linear'))
  Y=model.predict(X)
  plt.subplot(2,3,i+2)
  plt.scatter(Y[:,0],Y[:,1],c=color)
  plt.axis('off')
```

```
#visualizing linear transformations
fig=plt.figure()
plt.subplot(2,3,1)
plt.scatter(X[:,0],X[:,1],c=color)
plt.axis('off')
plt.title('Original')

for i in range(5):
  model=Sequential()
  model.add(Dense(2,input_dim=X.shape[1],activation='linear'))
  model.add(Dense(2,activation='tanh'))
  Y=model.predict(X)
  plt.subplot(2,3,i+2)
  plt.scatter(Y[:,0],Y[:,1],c=color)
  plt.axis('off')
```

# Visualizing Non-linear and Deep Transformations

```
fig=plt.figure()
fig.suptitle('2 layers+tanh')
plt.subplot(2,3,1)
plt.scatter(X[:,0],X[:,1],c=color)
plt.axis('off')
plt.title('Original')

for i in range(5):
  model=Sequential()
  model.add(Dense(2,input_dim=X.shape[1],activation='tanh'))
  model.add(Dense(2,activation='tanh'))
  Y=model.predict(X)
  plt.subplot(2,3,i+2)
  plt.scatter(Y[:,0],Y[:,1],c=color)
  plt.axis('off')

#visualizing linear transformations
fig=plt.figure()
fig.suptitle('3 layers+tanh')
plt.subplot(2,3,1)
plt.scatter(X[:,0],X[:,1],c=color)
plt.axis('off')
plt.title('Original')

for i in range(5):
  model=Sequential()
  model.add(Dense(2,input_dim=X.shape[1],activation='tanh'))
  model.add(Dense(2,activation='tanh'))
  model.add(Dense(2,activation='tanh'))
  Y=model.predict(X)
  plt.subplot(2,3,i+2)
  plt.scatter(Y[:,0],Y[:,1],c=color)
  plt.axis('off')
```
