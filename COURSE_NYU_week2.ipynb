```
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
#-------------------------------------------------------------------#
SMALL_SIZE = 12
MEDIUM_SIZE = 14
BIGGER_SIZE = 14

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE, titleweight='bold')     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE, labelweight='bold')    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
#-------------------------------------------------------------------#
 
```

```
X=np.linspace(-1,1,100).reshape(-1,1)
y=2*np.power(X,9)+1*np.power(X,2)+0.5*np.random.rand(X.shape[0],1)
plt.scatter(X,y)
plt.axis('equal');
```

# **2-Layer Linear Model**

```
neurons=100
lin=Sequential()
lin.add(Dense(neurons,input_dim=X.shape[1],activation='linear'))
lin.add(Dense(neurons,activation='linear'))
lin.add(Dense(1, activation='linear'))
lin.compile(loss='mean_squared_error', optimizer='adam',metrics=['mean_squared_error'])
lin.fit(X, y, epochs=1000, batch_size=32, verbose=0)

y_pred=lin.predict(X)
plt.scatter(X, y)
plt.plot(X, y_pred, 'r-', lw=5)
plt.title('linear model')
plt.axis('equal');
```

# **2-Layer Model with RELU**

```
neurons=100
relu=Sequential()
relu.add(Dense(neurons,input_dim=X.shape[1],activation='relu'))
relu.add(Dense(neurons,activation='relu'))
relu.add(Dense(1, activation='linear'))
relu.compile(loss='mean_squared_error', optimizer='adam',metrics=['mean_squared_error'])
relu.fit(X, y, epochs=1000, batch_size=32, verbose=0)

y_pred=relu.predict(X)
plt.scatter(X, y)
plt.plot(X, y_pred, 'r-', lw=5)
plt.title('relu')
plt.axis('equal');
```

# **2-Layer Model with Tanh**

```
neurons=100
tan=Sequential()
tan.add(Dense(neurons,input_dim=X.shape[1],activation='tanh'))
tan.add(Dense(neurons,activation='tanh'))
tan.add(Dense(1, activation='linear'))
tan.compile(loss='mean_squared_error', optimizer='adam',metrics=['mean_squared_error'])
tan.fit(X, y, epochs=1000, batch_size=32, verbose=0)

y_pred=tan.predict(X)
plt.scatter(X, y)
plt.plot(X, y_pred, 'r-', lw=5)
plt.title('tanh')
plt.axis('equal');
```

```
neurons=100
sig=Sequential()
sig.add(Dense(neurons,input_dim=X.shape[1],activation='sigmoid'))
sig.add(Dense(neurons,activation='sigmoid'))
sig.add(Dense(1, activation='linear'))
sig.compile(loss='mean_squared_error', optimizer='adam',metrics=['mean_squared_error'])
sig.fit(X, y, epochs=1000, batch_size=32, verbose=0)

y_pred=sig.predict(X)
plt.scatter(X, y)
plt.plot(X, y_pred, 'r-', lw=5)
plt.title('sigmoid')
plt.axis('equal');
```

# **Extrapolations**

```
X_new=np.linspace(-4,4,100).reshape(-1,1)
y_new=np.power(X,3)+np.random.rand(X.shape[0],1)
y_pred_lin=lin.predict(X_new)
y_pred_relu=relu.predict(X_new)
y_pred_tan=tan.predict(X_new)
y_pred_sig=sig.predict(X_new)

plt.scatter(X_new, y_new)
plt.plot(X_new, y_pred_lin, 'b-', lw=5,label='linear')
plt.plot(X_new, y_pred_relu, 'r-', lw=5,label='relu')
plt.plot(X_new, y_pred_tan, 'k-', lw=5,label='tanh')
plt.plot(X_new, y_pred_sig, 'g-', lw=5,label='sigmoid')

plt.legend()
plt.axis([-4,4,-10,10]);
```
